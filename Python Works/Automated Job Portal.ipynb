```python
for index, row in df.head(0).iterrows():
    # Reconnect every time
    server = smtplib.SMTP("smtp.gmail.com", 587)
    server.starttls()
    server.login("yashautomationthakur@gmail.com", "juxiqvfejkaixrly")

    # Prepare email body
    body = template.render(name=row["Name"], company=row["Company"])
    msg = MIMEText(body, "plain")
    msg["Subject"] = "Exploring Collaboration Opportunities"
    msg["From"] = "yashautomationthakur@gmail.com"
    msg["To"] = row["Email"]

    # Send email
    print(f"Sending to: {row['Name']} <{row['Email']}>")
    server.sendmail(msg["From"], [msg["To"]], msg.as_string())

    # Disconnect after each send
    server.quit()
```
